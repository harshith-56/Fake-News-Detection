```python
from collections import Counter
import re
def tokenize(text):
    text = re.sub(r'\W+', ' ', text.lower())
    return text.split()
all_words = df['text'].apply(tokenize).sum()
most_common_words = Counter(all_words).most_common(20)
print(most_common_words)
```

```python
df['label'] = df['label'].map({'FAKE': 0, 'REAL': 1})
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df[['text_length', 'title_length', 'label']].corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = df[['title', 'text']]
y = df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.34, random_state=36)
print("Training set shape (X_train):", X_train.shape)
print("Test set shape (X_test):", X_test.shape)
print("Training set labels shape (y_train):", y_train.shape)
print("Test set labels shape (y_test):", y_test.shape)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


def clean_text(text):
    text = re.sub(r'\W+', ' ', text.lower())
    return text
X_train['title'] = X_train['title'].apply(clean_text)
X_train['text'] = X_train['text'].apply(clean_text)
X_test['title'] = X_test['title'].apply(clean_text)
X_test['text'] = X_test['text'].apply(clean_text)

X_train_combined = X_train['title'] + " " + X_train['text']
X_test_combined = X_test['title'] + " " + X_test['text']

vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train_combined)
X_test_vectorized = vectorizer.transform(X_test_combined)
```

```python
print(X_train_combined)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score


model = LogisticRegression()

model.fit(X_train_vectorized, y_train)

y_pred = model.predict(X_test_vectorized)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB


nb_model = MultinomialNB()


nb_model.fit(X_train_vectorized, y_train)

nb_pred = nb_model.predict(X_test_vectorized)

print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_pred))
print(classification_report(y_test, nb_pred))
```

```python
from sklearn.svm import SVC


svm_model = SVC(kernel='linear')


svm_model.fit(X_train_vectorized, y_train)

svm_pred = svm_model.predict(X_test_vectorized)

print("SVM Accuracy:", accuracy_score(y_test, svm_pred))
print(classification_report(y_test, svm_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier


rf_model = RandomForestClassifier()

rf_model.fit(X_train_vectorized, y_train)

rf_pred = rf_model.predict(X_test_vectorized)

print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier


pac_model = PassiveAggressiveClassifier(C=0.1, tol=0.001, max_iter=1000, random_state=42)


pac_model.fit(X_train_vectorized, y_train)


pac_pred = pac_model.predict(X_test_vectorized)

print("Passive Aggressive Accuracy:", accuracy_score(y_test, pac_pred))
print(classification_report(y_test, pac_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import PassiveAggressiveClassifier


pac = PassiveAggressiveClassifier()

param_grid = {
    'C': [0.1, 1, 10, 100],
    'max_iter': [1000, 5000, 10000],
    'tol': [1e-3, 1e-4, 1e-5]
}


grid_search = GridSearchCV(pac, param_grid, cv=5, scoring='accuracy', n_jobs=-1)


grid_search.fit(X_train_vectorized, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best cross-validated accuracy:", grid_search.best_score_)
```

```python
from sklearn.model_selection import cross_val_score


best_model = grid_search.best_estimator_

cv_scores = cross_val_score(best_model, X_train_vectorized, y_train, cv=5)

print("Cross-validated scores:", cv_scores)
print("Mean CV accuracy:", cv_scores.mean())
```
